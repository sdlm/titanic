```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/var/tmp/output_file.csv')
df['day'] = df['timestamp'].apply(lambda x: int(str(x)[8:10]))
df['hour'] = df['timestamp'].apply(lambda x: int(x[11:13]))
df.head()
```

```python
columns_order = df.groupby('instrument').size().sort_values(ascending=False)
sns.countplot(y='instrument', order=columns_order.index, data=df);
```

```python
sns.countplot('period', data=df);
```

```python
sns.countplot(y='day', data=df);
```

```python
sns.distplot(df['hour'])
```
